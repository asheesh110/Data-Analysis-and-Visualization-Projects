```python
#loading important libraries

import pandas as pd
from matplotlib import pyplot as plt
```

```python
marvel = pd.read_csv('D:\\Python programs\\Datasets\\charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
#Find the categories of alignment

marvel['Alignment'].value_counts()
```

```python
#find the good alignment heros

good= marvel[marvel['Alignment']=='good']
good.head()
```

```python
#find the superheros having highest speed

good.sort_values(by=['Speed'],ascending=False)[0:5]
```

```python
#find the superheros having highest power

good.sort_values(by=['Power'],ascending=False)[0:5]
```

```python
good_max_power_full= good[good['Power']==100]
good_max_power_full.shape
good_max_power_full
```

```python
good_max_power=good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(good_max_power['Name'][0:5]),list(good_max_power['Total'][0:5]))
plt.show()
```

```python
bad = marvel[marvel['Alignment']=='bad']
```

```python
bad.head()
```

```python
#Finding  bad characters having highest power

bad.sort_values(by=['Power'],ascending=False)[0:5]
```

```python
#Finding  bad characters having highest intelligence

bad.sort_values(by=['Intelligence'],ascending=False)[0:5]
```

```python
#Finding  bad characters having highest total 

bad.sort_values(by=['Total'],ascending=False)[0:5]
```

```python
bad_max_power=bad.sort_values(by=['Total'],ascending=False)
bad_max_power.head()
```

```python
#making a histogram speed for good superheros
plt.figure(figsize=(10,7))
plt.hist(good['Speed'])
plt.title('Distribution for speed')
plt.xlabel('Speed')
plt.ylabel('Number of characters')
plt.show()
```

```python
#Making a histogram for combat for supervillans

plt.figure(figsize=(10,7))
plt.hist(bad['Combat'])
plt.xlabel('Combat')
plt.ylabel('Number of Supervillens')
plt.title("Distribution of combat")
plt.show()
```
