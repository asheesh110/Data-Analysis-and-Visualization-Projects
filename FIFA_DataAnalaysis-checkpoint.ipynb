```python
#making a barplot for the top five countries having most number of players belonging to that country
plt.figure(figsize=(7,7))
sns.barplot(x=list(fifa['nationality'].value_counts()[0:5].keys()),y=list(fifa['nationality'].value_counts()[0:5]))
plt.show()
```

```python
#extract two columns having player name and salary
player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
#arranging the player_salary dataset into descending order based on their salary 
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
#Making a bar plot for top five players getting highest salary
plt.figure(figsize=(8,8))
sns.barplot(x=list((player_salary['short_name'])[0:5]),y=list((player_salary['wage_eur'])[0:5]))
plt.show()
```
